```python
import matplotlib.pylab as plt
import numpy as np

# Points of x-axis
x = np.arange(0,10)

# Points of y-axis

y = np.arange(20, 30)
```

# Scatter

```python
plt.scatter(x,y,c='g')
```

```python
plt.scatter(x, y, c="b")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("This is the Title")
```

# Plot

```python
y = x*x
plt.plot(x,y)
```

```python
plt.plot(x,y, "r--")
```

```python

plt.plot(x,y, "o")
```

```python
plt.plot(x,y, "o--")
```

```python

plt.plot(x,y, "r*")
```

```python
plt.plot(x,y, "r*", linestyle="dashed",lineWidth=2, markerSize=12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

# SupPlot


```python
# subplot(number_of_rows, number_of_cols, index)
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,4)
plt.plot(x,y,'b')
```

# sin wave form


```python
x = np.arange(0, 4* np.pi, 0.2)

y = np.sin(x)

plt.plot(x, y)
```

```python
x
```

```python
x = np.arange(0, 6* np.pi, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")
```

# Bar plot


```python
x1 = [2,5,9]
y1 = [10,15,20]

x2 = [7,10,16]
y2 = [8,15,10]
plt.bar(x1,y1, color = 'g')
plt.bar(x2,y2, color = 'b')
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

# Pie Charts

```python
arr = [35, 15, 23, 27]
plt.pie(arr)
```

```python
arr_label = ["Python", 'Java','C','C++']
plt.pie(arr, labels=arr_label)
```

```python
color= ['green','gold','purple',"lightcoral"]
explode = [0.1, 0, 0, 0]

plt.pie(arr, labels=arr_label, colors = color, explode=explode, autopct='%1.1f%%',shadow = True)
```
