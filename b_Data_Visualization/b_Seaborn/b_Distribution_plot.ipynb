# Distribution plot

```python
import seaborn as sns

dataset = sns.load_dataset('tips') # tips is the in memory dataset
dataset.head()
```

### Compute pairwise correlation of columns, excluding NA/null values.
>Correlation is a statistical measure that expresses the extent to which two variables are linearly related
> (meaning they change together at a constant rate). It's a common tool for describing simple relationships without making a statement about cause and effect.

```python
dataset.corr()
```

### heatmap()
> Plot rectangular data as a color-encoded matrix.

> A heatmap is a two-dimensional graphical representation of data where the individual values that are contained
> in a matrix are represented as colors.

```python
sns.heatmap(dataset.corr())
```

# Jointplot
>Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two
> variables and describe their individual distributions on the same plot.

```python
sns.jointplot(x='tip',y='total_bill', data=dataset, kind='hex')
```

##  jointplot- kind
>kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
> reg - > regression
> kde -> kernel density

> You can see more about jointplot here,
> http://alanpryorjr.com/visualizations/seaborn/jointplot/jointplot/

```python
sns.jointplot(x='tip',y='total_bill', data=dataset, kind='reg')
```

```python
# by default kind is scatter
sns.jointplot(x='tip',y='total_bill', data=dataset)
```
